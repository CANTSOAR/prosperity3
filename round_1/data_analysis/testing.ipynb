```python
N = 500

bid_out_sample_kelp = p2[p2["product"] == "KELP"]["bid_price_1"][-N:]
bid_out_sample_ink = p2[p2["product"] == "SQUID_INK"]["bid_price_1"][-N:]
ask_out_sample_kelp = p2[p2["product"] == "KELP"]["ask_price_1"][-N:]
ask_out_sample_ink = p2[p2["product"] == "SQUID_INK"]["ask_price_1"][-N:]

from sklearn.linear_model import LinearRegression  # or try RandomForest, MLP, etc.

X, y_diff = Regressor.arima_make_dataset(bid_out_sample_kelp.values, bid_order_kelp)
bid_model_kelp = LinearRegression().fit(X, y_diff)
Regressor.eval_model(bid_out_sample_kelp.values, bid_model_kelp, order = bid_order_kelp, steps = 1)

forecast_kelp = Regressor.forecast(bid_out_sample_kelp.values, bid_model_kelp, order = bid_order_kelp, steps = 100)

X, y_diff = Regressor.arima_make_dataset(bid_out_sample_ink.values, bid_order_ink)
bid_model_ink = LinearRegression().fit(X, y_diff)
Regressor.eval_model(bid_out_sample_ink.values, bid_model_ink, order = bid_order_ink, steps = 1)

forecast_kelp = Regressor.forecast(bid_out_sample_kelp.values, bid_model_ink, order = bid_order_ink, steps = 100)

X, y_diff = Regressor.arima_make_dataset(ask_out_sample_kelp.values, ask_order_kelp)
ask_model_kelp = LinearRegression().fit(X, y_diff)
Regressor.eval_model(ask_out_sample_kelp.values, ask_model_kelp, order = ask_order_kelp, steps = 1)

forecast_kelp = Regressor.forecast(ask_out_sample_kelp.values, ask_model_kelp, order = ask_order_kelp, steps = 100)

X, y_diff = Regressor.arima_make_dataset(ask_out_sample_ink.values, ask_order_ink)
ask_model_ink = LinearRegression().fit(X, y_diff)
Regressor.eval_model(ask_out_sample_ink.values, ask_model_ink, order = ask_order_ink, steps = 1)

forecast_kelp = Regressor.forecast(ask_out_sample_kelp.values, ask_model_ink, order = ask_order_ink, steps = 100)

print("Bid Kelp Weights:", bid_model_kelp.coef_)      # array of weights
print("Bid Kelp Intercept:", bid_model_kelp.intercept_)
print("Bid Ink Weights:", bid_model_ink.coef_)      # array of weights
print("Bid Ink Intercept:", bid_model_ink.intercept_)
print("Ask Kelp Weights:", ask_model_kelp.coef_)      # array of weights
print("Ask Kelp Intercept:", ask_model_kelp.intercept_)
print("Ask Ink Weights:", ask_model_ink.coef_)      # array of weights
print("Ask Ink Intercept:", ask_model_ink.intercept_)
```

```python
kelp_bid_forecasts = [bid_out_sample_kelp.values[0]]
last_val = kelp_bid_forecasts[-1]

for value in bid_out_sample_kelp.values[1:]:
    if value == last_val:
        kelp_bid_forecasts.append(value)
    else:
        kelp_bid_forecasts.append(kelp_bid_forecasts[-1])
        last_val = value

kelp_ask_forecasts = [ask_out_sample_kelp.values[0]]
last_val = kelp_ask_forecasts[-1]

for value in ask_out_sample_kelp.values[1:]:
    if value == last_val:
        kelp_ask_forecasts.append(value)
    else:
        kelp_ask_forecasts.append(kelp_ask_forecasts[-1])
        last_val = value
```

```python
kelp_ask_forecasts = pd.DataFrame(kelp_ask_forecasts)
kelp_ask_forecasts.index = bid_out_sample_kelp.index
kelp_bid_forecasts = pd.DataFrame(kelp_bid_forecasts)
kelp_bid_forecasts.index = bid_out_sample_kelp.index
```
